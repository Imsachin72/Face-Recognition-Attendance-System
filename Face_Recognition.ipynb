```python
import cv2
import face_recognition
import csv
from datetime import datetime
```

```python
# Load the known face image
known_image1 = face_recognition.load_image_file("C:\\Users\\KALI LINNUX\\Desktop\\Python(coding thinker)\\Itself\\face_recognition\\face\\rajdeep.jpg")
known_image2 = face_recognition.load_image_file("C:\\Users\\KALI LINNUX\\Desktop\\Python(coding thinker)\\Itself\\face_recognition\\face\\Gaurav.jpg")
known_image3 = face_recognition.load_image_file("C:\\Users\\KALI LINNUX\\Desktop\\Python(coding thinker)\\Itself\\face_recognition\\face\\Abhay_Lodhi.png")
known_image4 = face_recognition.load_image_file("C:\\Users\\KALI LINNUX\\Desktop\\Python(coding thinker)\\Itself\\face_recognition\\face\\Atul_saxna.png")
known_image5 = face_recognition.load_image_file("C:\\Users\\KALI LINNUX\\Desktop\\Python(coding thinker)\\Itself\\face_recognition\\face\\Aks.png")
known_image6 = face_recognition.load_image_file("C:\\Users\\KALI LINNUX\\Desktop\\Python(coding thinker)\\Itself\\face_recognition\\face\\abhijeet.png")

# Encode the known face
known_face_encoding1 = face_recognition.face_encodings(known_image1)[0]
known_face_encoding2 = face_recognition.face_encodings(known_image2)[0]
known_face_encoding3 = face_recognition.face_encodings(known_image3)[0]
known_face_encoding4 = face_recognition.face_encodings(known_image4)[0]
known_face_encoding5 = face_recognition.face_encodings(known_image5)[0]
known_face_encoding6 = face_recognition.face_encodings(known_image6)[0]


# creating arrays of know face encoding and their names
known_face_encodings = [
                    known_face_encoding1,
                    known_face_encoding2,
                    known_face_encoding3,
                    known_face_encoding4,
                    known_face_encoding5,
                    known_face_encoding6
                    ]

know_face_names = [ 
                    'Rajdeep Namdev', 
                    'Gaurav singh Bhandari',
                    'Abhay Lodhi',
                    'Atul Saxna',
                    'Akshatraj Singh Tomar',
                    'Abhijeet Sen'
                 ]

# Initialize some variables
face_locations = []
face_encodings = []
face_names = []
present_students = set()

# load present date and time
now = datetime.now()
today = now.day
month = now.month
year = now.year

# Load attendance data from CSV if available
attendance_file_path = "Student_Attendance.csv"

try:
    with open(attendance_file_path, mode='r') as file:
        reader = csv.reader(file)
        for row in reader:
            present_students.add(row[0])
except FileNotFoundError:
    pass

# Capture video from webcam
video_capture = cv2.VideoCapture(0)

while True:
    # Capture frame-by-frame
    ret, frame = video_capture.read()

    # Find all the faces and face encodings in the current frame
    face_locations = face_recognition.face_locations(frame)
    face_encodings = face_recognition.face_encodings(frame, face_locations)

    face_names = []
    for face_encoding in face_encodings:
        # Check if the face matches the known face
        matches = face_recognition.compare_faces(known_face_encodings, face_encoding)
        name = "Unknown_student"

        if True in matches:
            first_match_index = matches.index(True)
            name = know_face_names[first_match_index]
            present_students.add(name)

        face_names.append(name)

    # Display the results
    for (top, right, bottom, left), name in zip(face_locations, face_names):
        
        # Draw a rectangle around the face
        cv2.rectangle(frame, (left, top), (right, bottom), (0, 0, 255), 2)

        # Draw a label with a name below the face
        cv2.rectangle(frame, (left, bottom - 30), (right, bottom), (0, 0, 255), cv2.FILLED)
        
        font = cv2.FONT_HERSHEY_DUPLEX
        cv2.putText(frame, name, (left + 5, bottom - 5), font, 0.5, (255, 255, 255), 1)

    # Display the resulting image
    cv2.imshow('Video', frame)

    # Exit loop if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Save the attendance to CSV
with open(attendance_file_path, mode='w', newline='') as file:
    writer = csv.writer(file)
    for student in present_students:
        writer.writerow([student, f"{today}-{month}-{year}", 'present'])

# Release handle to the webcam
video_capture.release()
cv2.destroyAllWindows()
```
